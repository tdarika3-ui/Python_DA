**10 OUTLIER DETECTION AND HANDLING**

**FIND OUTLIERS USING IQR METHOD**

```python
Q1=df['Salary'].quantile(0.25)
Q3=df['Salary'].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
outliers=df[(df['Salary']<lower)|(df['Salary']>upper)]
print(outliers)
```

REMOVE OUTLIERS USING IQR

```python
Q1=df['Salary'].quantile(0.25)
Q3=df['Salary'].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
df_cleaned_outliers=df[(df['Salary']>=lower)&(df['Salary']<=upper)]
print(df_cleaned_outliers)
```

**FIND OUTLIERS USING Z-SCORE METHOD**

```python
df['Z_Score']=zscore(df['Salary'])
df[df['Z_Score']>3]
```

```python
#create zscore column
df['Z_Score']=zscore(df['Salary'])
df_cleaned=df[(df['Salary']<3)|(df['Salary']>-3) ]
print(df_cleaned)
```
